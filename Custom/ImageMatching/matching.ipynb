```python
# Import all libraries here
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def MatchImage(testImg, matchImage):
    imageMatch = False
    img_gray = cv2.cvtColor(testImg, cv2.COLOR_BGR2GRAY)
    w, h = matchImage.shape[::-1]
    res = cv2.matchTemplate(img_gray,matchImage,cv2.TM_CCOEFF_NORMED)
    loc = np.where( res >= threshold)
    for pt in zip(*loc[::-1]):
        cv2.rectangle(testImg, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
        imageMatch = True
    if imageMatch == True :
        plt.imshow(testImg)
        plt.show()
    return imageMatch
```

```python

iteration = 8
i = 1
img_rgb = cv2.imread('images/test4.jpg')
plt.imshow(img_rgb)
plt.show()


while i <= iteration:
    template = cv2.imread('images/canola' + str(i) + '.jpg',0)
    threshold = 0.8
    if MatchImage(img_rgb, template) == True :
        template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
        plt.imshow(template)
        plt.show()
    i = i + 1

#cv2.imwrite('res.png',img_rgb)
```
